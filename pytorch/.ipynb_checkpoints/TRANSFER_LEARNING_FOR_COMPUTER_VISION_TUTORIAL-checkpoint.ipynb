本チュートリアルでは、転移学習でどのように画像分類用のCNNを学習させるかを学びます。  
転移学習について知りたい方は、[こちら](https://cs231n.github.io/transfer-learning/)をご確認ください。

上から引用すると、「CNNをスクラッチで学習させる人は少ないです。データセットが少ないからです。ネットワークをとても大きなサイズのデータセット（e.g. ImageNet, 1000カテゴリの120万個の画像を含む）で学習させておくのが一般的です。その後、ネットワークを特徴量抽出器として扱うのがよくされます。」

転移学習の方法としては以下の2種類がメジャーです。  
<ul>
    <li>finetuning: 学習済みのネットワークを使いますが、重みを改めて学習させます。</li>
    <li>feature extraction: 最後の全結合層以外の重みを固定します。最後の層だけ学習させます。</li>
</ul>

```python
# License: BSD
# Author: Sasank Chilamkurthy

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import numpy as np
import torchvision
from torchvision import datasets,models,transforms
import matplotlib.pyplot as plt
import time
import os
import copy

plt.ion() # インタラクティブモード
```

# データのロード

データのロードにはtorchvisionとtorch.utils.dataを使います。  
このチュートリアルでは、antsとbeesを識別するモデルを学習させます。それぞれ120枚の学習用画像、75枚の検証用画像を持っています。  
もしスクラッチで学習させるなら、データセットは小さすぎます。  
転移学習の出番です。

まずはデータセットをダウンロードします。

```python
# # wgetでデータを取得し、unzipする
# ! wget https://download.pytorch.org/tutorial/hymenoptera_data.zip -P ./../data/
# ! unzip ./../data/hymenoptera_data.zip -d ./../data/
```

```python
# 学習ではdata augmentationと正規化
# 検証では正規化のみ
data_transforms={
    "train":transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    "val":transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
}

data_dir="./../data/hymenoptera_data/" # データセットのディレクトリ 
image_datasets={x:datasets.ImageFolder(os.path.join(data_dir,x),data_transforms[x]) for x in ["train","val"]} # 画像のデータセット
dataloaders={x:torch.utils.data.DataLoader(image_datasets[x],batch_size=4,shuffle=True,num_workers=4) for x in ["train","val"]} # データローダ
dataset_sizes={x:len(image_datasets[x]) for x in ["train","val"]} # データセットのサイズ
class_names=image_datasets["train"].classes # クラス

device=torch.device("cuda:0" if torch.cuda.is_available() else "cpu") # デバイス
```

# 画像の確認

画像を確認しましょう。

```python
def imshow(inp,title=None): # 画像を表示する関数
    # テンソルを入力とする
    inp=inp.numpy().transpose((1,2,0))
    mean=np.array([0.485, 0.456, 0.406]) # 平均
    std=np.array([0.229, 0.224, 0.225]) # 標準偏差
    inp=std*inp+mean # 正規化されたものを、非正規化
    plt.imshow(inp) # 画像を表示
    
    if title is not None: # タイトルが存在する場合
        plt.title(title) # タイトルも表示
        
    plt.pause(0.001) # 画像が更新されるのを待つ

inputs,classes=next(iter(dataloaders["train"])) # 学習用データセットのバッチを取得
out=torchvision.utils.make_grid(inputs) # 画像をグリッドに配置

imshow(out,title=[class_names[x] for x in classes]) # 画像を表示
```

# モデルの学習

## 関数の定義

まず以下の動作をする、モデルを学習させる関数を定義します。  
<ul>
    <li>学習率をスケジュールする</li>
    <li>ベストなモデルをセーブする</li>
</ul>
以下では、パラメータschedulerは、torch.optim.lr_schedulerの学習率スケジューラオブジェクトです。

```python
def train_model(model,criterion,optimizer,scheduler,num_epochs=25):
    since=time.time() # 時間を計測開始
    best_model_wts=copy.deepcopy(model.state_dict())
    best_acc=0.0 # これまでのベストの精度
    
    for epoch in range(num_epochs):
        print("Epoch {}/{}".format(epoch,num_epochs-1))
        print("-"*10)
        
        # 各エポックには学習と検証フェーズがある
        for phase in ["train","val"]:
            if phase=="train": # 訓練フェーズでは
                model.train() # 訓練モード
            else:
                model.eval() # 検証モード
                
            running_loss=0.0 # ロス
            running_corrects=0 # 分類が正しいものの数のカウント
            
            # 各データローダのバッチで繰り返す
            for inputs,labels in dataloaders[phase]:
                inputs=inputs.to(device) # 入力
                labels=labels.to(device) # ラベル
                
                optimizer.zero_grad() # 勾配をゼロにする
                
                with torch.set_grad_enabled(phase=="train"):
                    outputs=model(inputs) # モデルの出力
                    _,preds=torch.max(outputs,1) # 予測値が最大のクラスを取得
                    loss=criterion(outputs,labels) # ロスを計算
                    
                    if phase=="train": # 学習フェーズのときだけ、パラメータを最適化する
                        loss.backward() # 逆伝搬
                        optimizer.step() # 最適化
                        
                running_loss+=loss.item()*inputs.size(0) # ロスを計算する
                running_corrects+=torch.sum(preds==labels.data) # 分類が正しかったもののカウントを増やす
                
            if phase=="train": # 学習フェイズのとき
                scheduler.step() # 最適化する
                
            ### エポックでの予測性能を見る
            epoch_loss=running_loss/dataset_sizes[phase] # エポックでのロス
            epoch_acc=running_corrects.double()/dataset_sizes[phase] # エポックでの正しく分類できた数
            
            print("{} Loss: {:.4f} Acc: {:.4f}".format(phase,epoch_loss,epoch_acc))
            
            ### 精度が過去最高の場合、モデルをdeep copyする
            if phase=="val" and epoch_acc>best_acc:
                best_acc=epoch_acc # 精度を更新
                best_model_wts=copy.deepcopy(model.state_dict()) # ベストなモデルの重み
                
        print()
        
    time_elapsed=time.time()-since # 経過時間
    print("Training complete in {:.0f}m {:.0f}s".format(time_elapsed//60,time_elapsed%60))
    print("Best val Acc: {:.4f}".format(best_acc))
    
    model.load_state_dict(best_model_wts) # モデルの重みをロード
    
    return model
```

## 画像を表示する関数

画像何個かを表示する関数です。

```python
def visualize_model(model,num_images=6):
    was_training=model.training # この関数を実行する直前のモード
    model.eval() # モデルを評価モードに
    images_so_far=0 # これまでに表示してきた画像の数
    
    with torch.no_grad(): # この中では勾配の更新はなし
        for i,(inputs,labels) in enumerate(dataloaders["val"]): # 検証用データの中で
            inputs=inputs.to(device) # 入力
            labels=labels.to(device) # ラベル
            outputs=model(inputs) # モデルの出力
            _,preds=torch.max(outputs,1) # 予測値の最大のクラス
            
            for j in range(inputs.size()[0]): # 各画像について
                images_so_far+=1 # これまでに見てきた画像の数
                ax=plt.subplot(num_images//2,2,images_so_far) # 軸
                ax.axis("off")
                ax.set_title("predicted: {}".format(class_names[preds[j]]))
                imshow(inputs.cpu().data[j]) # 画像の表示
                
                if images_so_far==num_images: # もし指定した数の画像を表示したら
                    model.train(mode=was_training) # モードを、この関数を実行する前に戻す
                    
                    return
                
        model.train(mode=was_training) # ?
```

## finetuning

学習済みのモデルを読み込み、最後の全結合層の重みをリセットします。

```python
model_ft=models.resnet18(pretrained=True) # 学習済みのresnetをダウンロードし、読み込む
num_ftrs=model_ft.fc.in_features # 特徴量の数

# 今回の出力のサイズは2なので下では2と設定しているが、一般化するにはlen(class_names)
model_ft.fc=nn.Linear(num_ftrs,2) # 最後の結合層
model_ft=model_ft.to(device) # モデルをデバイスに送る
criterion=nn.CrossEntropyLoss() # クロスエントロピー誤差

optimier_ft=optim.SGD(model_ft.parameters(),lr=0.001,momentum=0.9) # 最適化関数
exp_lr_scheduler=lr_scheduler.StepLR(optimier_ft,step_size=7,gamma=0.1) # 学習率のスケジューラ
```

## モデルの学習・検証

いよいよモデルを学習させます。

```python
### モデルの学習
model_ft=train_model(
    model=model_ft,
    criterion=criterion,
    optimizer=optimier_ft,
    scheduler=exp_lr_scheduler,
    num_epochs=25
)
```

```python
### 画像の表示
visualize_model(model=model_ft)
```

# feature extraction

ここでは、最終層以外の重みを固定します。そうするには、requires_grad=Falseとして、backward()で重みを更新しないようにします。

```python
model_conv=torchvision.models.resnet18(pretrained=True) # モデルの読み込み

for param in model_conv.parameters(): # 各パラメータについて
    param.reqiures_grad=False # 更新しないように設定
    
num_ftrs=model_conv.fc.in_features # 特徴量の数　
model_conv.fc=nn.Linear(num_ftrs,2) # 最終層を全結合層にする
model_conv=model_conv.to(device) # 　モデルをデバイスに送る
criterion=nn.CrossEntropyLoss() # クロスエントロピー誤差

optimizer_conv=optim.SGD(model_conv.fc.parameters(),lr=0.001,momentum=0.9) # 最適化関数

exp_lr_scheduler=lr_scheduler.StepLR(optimizer=optimizer_conv,step_size=7,gamma=0.1) # 各エポックで、学習率を0.1倍してゆく
```

## 学習・検証

それでは最終層のみを学習させます。

```python
model_conv=train_model(
    model=model_conv,
    criterion=criterion,
    optimizer=optimizer_conv,
    scheduler=exp_lr_scheduler,
    num_epochs=25
)
```

```python
visualize_model(model_conv) # 画像を表示

plt.ioff()
plt.show()
```

確かにいい感じに分類できていそうです。
