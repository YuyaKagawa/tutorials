# AUTOGRAD: AUTOMATIC DIFFERENTIATION

参考: https://pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html#sphx-glr-beginner-blitz-autograd-tutorial-py

PyTorchのニューラルネットワークの核となるのがautogradパッケージです。  
まずはこれを見てみましょう。 
autogradパッケージには自動微分の操作が入っています。define-by-runのフレームワークですので、逆伝搬はどのようにコードを実行したかに依存します。毎回変わります。

torch.Tensorがこのパッケージの中心となるクラスです。.requires_grad()をTrueにすることで、これからの操作を受け付けます。  
計算がすんだ後に.backward()とすることで自動的に全ての勾配を計算します。テンソルに対する勾配は.gradに蓄積されます。  
テンソルの操作をやめるときは、.detach()でできます。  
メモリを使いたくなかったり一時的に止めるときには、with torch.no_grad()でできます。モデルを評価するときにはこれは便利です。  
もう一つ、Functionというクラスが重要です。  
TensorとFunctionは相互作用しています。各テンソルには.grad_fn()があり、これはTensorを作ったFunctionを表しています。（grad_fn=Noneとしない限り）  
微分を計算したい場合は、Tensorで.backward()を計算します。もしTensorがスカラー（要素が1個）の場合は、backward()の引数を指定する必要はありません。  
もし要素が2つ以上ある場合は、適切な形を持ったgradientを指定する必要があります。

```python
import torch
```

テンソルを作り、requires_grad=Trueとします。

```python
x=torch.ones(2,2,requires_grad=True)
print(x)
```

テンソルの計算をします。

```python
y=x+2
print(y)
```

yは計算の結果生み出されたものなので、grad_fnを持っています。

```python
print(x.grad_fn)
print(y.grad_fn)
```

さらにyを計算してみましょう。

```python
z=y*y*3
out=z.mean()

print(z)
print(out)
```

.requires_grad_()でrequires_gradのフラグをinplaceで変更できます。デフォルトではFalseです。

```python
a=torch.randn(2,2)
a=((a*3)/(a-1))
print(a.requires_grad)

a.requires_grad_(True)
print(a.requires_grad)
b=(a*a).sum()
print(b.grad_fn)
```

では逆伝搬させてみましょう。ここではoutは1要素なので、out.backward()はout.backward(torch.tensor(1.))と同じです。

```python
out.backward()
```

勾配　d(out)/dx　を表示します。

```python
print(x.grad)
```

torch.autogradはvector-Jacobianの積を計算します。  
例を見てみましょう。

```python
x=torch.randn(3,requires_grad=True)
y=x*2

print(x)
print(y)

while y.data.norm()<1000:
    y=y*2
    print(y)
```

このケースではyはスカラーではありません。torch.autogradはJacobianを直接計算はできませんが、もしJacobianの積を得たかったら、backwardの引数に渡します。

```python
v=torch.tensor([0.1,1.0,0.0001],dtype=torch.float) 
y.backward(v,retain_graph=True)
print(x.grad)

v=torch.tensor([1,1,1],dtype=torch.float) 
y.backward(v,retain_graph=True)
print(x.grad)
```

autogradもwith torch.no_grad()内では止めることができます。

```python
print(x.requires_grad)
print((x**2).requires_grad)

with torch.no_grad():
    print((x**2).requires_grad)
```

もしくはdetach()を使って勾配情報だけ必要としないものを得ることができます。

```python
print(x.requires_grad)
y=x.detach()
print(y.requires_grad)
print(x.eq(y).all()) # eq()は各要素が一致するかを見る
```

# NEURAL NETWORKS

参考: https://pytorch.org/tutorials/beginner/blitz/neural_networks_tutorial.html#sphx-glr-beginner-blitz-neural-networks-tutorial-py

torch.nnパッケージを使うことでニューラルネットワークを構築することができます。このtorch.nnはこれまで見てきたautogradに依存しています。nn.Moduleにはレイヤーやメソッドforward(input)が含まれています。  
以降は下のネットワークを例に話を進めてゆきます。

```python
from IPython.display import Image,display_png
display_png(Image("./mnist.png"))
```

シンプルなfeed-forwardネットワークです。入力を受け、レイヤを通って、最後に出力します。ニューラルネットワークの学習プロセスは以下の通りです。  
<ol>
    <li>学習パラメータ（重み）を持つニューラルネットワークを定義する</li>
    <li>入力のデータセットについて繰り返す</li>
    <li>ネットワークを通して入力を進める</li>
    <li>ロスを計算する</li>
    <li>ネットワークのパラメータに勾配を逆伝搬させる</li>
    <li>ネットワークの重みを更新する</li>
</ol>

## ネットワークの定義

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        # 入力チャンネル数1、出力チャンネル6、カーネルは3*3
        self.conv1=nn.Conv2d(1,6,3)
        self.conv2=nn.Conv2d(6,16,3)
        
        # アフィン変換 y=Wx+b
        self.fc1=nn.Linear(16*6*6,120) # 6*6の画像の次元
        self.fc2=nn.Linear(120,84)
        self.fc3=nn.Linear(84,10)
        
    def forward(self,x):
        # (2,2)のウィンドウでMaxPooling
        x=F.max_pool2d(F.relu(self.conv1(x)),(2,2))
        x=F.max_pool2d(F.relu(self.conv2(x)),2) # 正方形の場合はスカラーでもOK
        x=x.view(-1,self.num_flat_features(x))
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.fc3(x)
        
        return x
    
    def num_flat_features(self,x):
        size=x.size()[1:] # バッチ以外の全次元
        num_features=1
        
        for s in size:
            num_features*=s
                        
        return num_features
```

```python
net=Net()
print(net)
```

forwardを定義する必要があります。backwardに関しては、autogradを使えば自動的に計算します。forwardではテンソルの操作ができます。  
学習可能なパラメータはnet.parameters()で取得できます。

```python
params=list(net.parameters())
print(len(params))
print(params[0].size()) # params[0].shapeでもよい
```

試しにランダムな32×32の入力を入れてみます。  
ここでは(LeNet)の入力サイズは32×32を想定しています。MNISTデータセットで使うためには、画像を32×32のサイズに調整してください。

```python
input=torch.randn(1,1,32,32)
out=net(input)
print(out)
```

全てのパラメータの勾配のバッファを0に初期化して、ランダムな勾配で逆伝搬をします。

```python
net.zero_grad()
out.backward(torch.randn(1,10))
```

torch.nnはミニバッチのみサポートしています。torch.nnのパッケージは1個のサンプルではなく、ミニバッチだけを入力として扱えます。  
例えば、nn.Conv2dは（サンプル数×チャンネル数×高さ×幅）の4次元のテンソルを入力とできます。  
サンプルサイズが1の場合、input.unsqueeze(0)をしてフェイクのバッチの次元を追加してください。

torch.Tensor: backward()などのautogradの操作ができる、多次元配列。テンソルの勾配も持っている。  
nn.Module: ニューラルネットワークのモジュール。GPUに移動させたり、出力したりといった操作に便利。  
nn.Parameter: パラメータのテンソル。Moduleの属性として自動的に割り当てられる。  
autograd.Function: autogradのforwardとbackwardを実行する。テンソルの操作は少なくとも1個のFunctionノードを生成する。

ここまで、  
<ul>
    <li>ニューラルネットワークを定義する</li>
    <li>入力を進め、逆伝搬を実行する</li>
</ul>
ところを見てきました。  
まだ、
<ul>
    <li>ロスを計算する</li>
    <li>ネットワークの重みを更新する</li>
</ul>
ことをしていません。

## ロス関数

ロス関数は入力の(output,target)を受け、ターゲットから出力がどの程度離れているかを計算します。  
nnパッケージには様々なロス関数があります。シンプルなのは、nn.MSELossです。  

```python
output=net(input) # ネットワークの出力を計算
target=torch.randn(10) # ダミーのtarget
target=target.view(1,-1) # outputと同じ形にする
criterion=nn.MSELoss()

loss=criterion(output,target) # ロスを計算する
print(loss)
```

計算グラフは下のような感じです。

```python
display_png(Image("./computation_graph.png"))
```

なので、loss.backward()を実行するときに、全グラフがロスという点で微分され、requires_grad=Trueとした全テンソルは.gradのテンソルを持つことになります。

```python
print(loss.grad_fn)
print(loss.grad_fn.next_functions)
print(loss.grad_fn.next_functions[0][0])
print(loss.grad_fn.next_functions[0][0].next_functions[0][0]) # ReLU
```

## 逆伝搬

loss.backward()で逆伝搬させることができます。勾配は加算されてゆくので、最初に初期化する必要があります。  
loss.backward()を使い、backward()の前後でconv1の勾配がどう変化したのかを見ます。

```python
net.zero_grad() # 全てのパラメータの勾配のバッファをゼロにする

print("conv1.bias.grad before backward")
print(net.conv1.bias.grad)

loss.backward()

print("conv1.bias.grad after backward")
print(net.conv1.bias.grad)
```

ここまでで、ロス関数の使い方を見てきました。

それではネットワークの重みを更新します。

## 重みの更新

ここでは一番シンプルな、SGDで重みを更新することとします。

```python
learning_rate=0.01 # 学習率

for f in net.parameters():
    f.data.sub_(f.grad.data*learning_rate)
```

torch.optimを使うことでSGD以外にもAdamとかRMSPropとかあります。

```python
import torch.optim as optim

# 最適化関数を自作する
optimizer=optim.SGD(net.parameters(),lr=0.01)

# 学習フェーズ
optimizer.zero_grad() # 勾配のバッファを0にする
output=net(input)
loss=criterion(output,target)
loss.backward()
optimizer.step() # 更新する
```

```python
print(net.conv1.bias.grad)
```

### CIFAR10のデータセットの読み込み・正規化

torchvisionを使うと簡単にCIFAR10を読み込むことができます。

```python
import torch
import torchvision
import torchvision.transforms as transforms
```

torchvisionのデータセットの出力はPILImageで、値の範囲は[0,1]です。それを正規化して範囲を[-1,1]としたテンソルに変換します。

```python
trainset=torchvision.datasets.CIFAR10(
    root="./data",
    train=True, # 訓練する
    download=True, # ダウンロードする
    transform=transform #  PIL imageを入力として値を出力する
)

testset=torchvision.datasets.CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=transform
)

classes=(
    "plane","car","bird","cat","deer",
    "dog","frog","horse","ship","truck"
)
```

データを見てみます。

```python
print(trainset.data.shape)
print(trainset.data[0,:,:,0]) # 1個目の画像の1チャンネル目の値
print(trainset.classes)
print(testset.data.shape)
print(testset.classes)
```

データローダも定義します。

```python
transform=transforms.Compose(
    [transforms.ToTensor(), # torch.Tensorに変換
     transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5)) # 正規化
    ]
)

trainloader=torch.utils.data.DataLoader(
    trainset,
    batch_size=4,
    shuffle=True,
    num_workers=1 # 何個のサブプロセスでデータのロードをするか
)

testloader=torch.utils.data.DataLoader(
    testset,
    batch_size=4,
    shuffle=False,
    num_workers=1
)

dataiter=iter(trainloader)
```

画像を表示してみましょう。

```python
import matplotlib.pyplot as plt
import numpy as np

# 画像を表示する関数を自作
def imshow(img):
    img=img/2+0.5 # [-1,1]を[0,1]に変換
    npimg=img.numpy() # numpyに変換する
    plt.imshow(np.transpose(npimg,(1,2,0)))
    plt.show()

images,labels=dataiter.next() # バッチサイズ分のデータを読み込む
imshow(torchvision.utils.make_grid(images)) # 画像を表示する
print(" ".join("%5s"%classes[labels[j]] for j in range(4))) # ラベルを表示する
```

## CNNの定義

前回までのニューラルネットワークを、1チャンネルではなく3チャンネル分に変更します。

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.conv1=nn.Conv2d(3,6,5) # 前回(1,6,3)
        self.pool=nn.MaxPool2d(2,2) #  前回なし
        self.conv2=nn.Conv2d(6,16,5) # 前回(6,16,3)
        self.fc1=nn.Linear(16*5*5,120) # 前回(16*6*6)
        self.fc2=nn.Linear(120,84) # 前回(120,84)
        self.fc3=nn.Linear(84,10) # 前回(84,10)
    
    def forward(self,x):
        x=self.pool(F.relu(self.conv1(x)))
        x=self.pool(F.relu(self.conv2(x)))
        x=x.view(-1,16*5*5)
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.fc3(x)
        
        return x
```

```python
net=Net()
```

## ロス関数と最適化の定義

クロスエントロピー誤差とmomentumありのSGDを使いましょう。

```python
import torch.optim as optim

criterion=nn.CrossEntropyLoss() # クロスエントロピー誤差
optimizer=optim.SGD(net.parameters(),lr=0.001,momentum=0.9) # SGD
```

## ネットワークの学習

ここからがマグマなんです。data iteratorをループして、ネットワークに入力して最適化させてゆきます。

```python
for epoch in range(2): # データセットについてループしてゆく
    running_loss=0.0 # ロス
    
    for i,data in enumerate(trainloader,0): # ミニバッチごとに
        inputs,labels=data # データは[inputs,labels]のリスト
        optimizer.zero_grad() # パラメータの勾配をゼロにする

        outputs=net(inputs) # forward+backward+optimize
        loss=criterion(outputs,labels) # ロスを計算する
        loss.backward()
        optimizer.step()

        running_loss+=loss.item() # ロスを足してゆく

        # 結果を表示する
        if i%2000==1999: # ミニバッチ2000個ごとに表示
            print("[%d,%5d] loss: %.3f"%(epoch+1,i+1,running_loss/2000))
            running_loss=0.0
        
print("Finished Training")
```

学習結果のモデルを保存しましょう。

## テスト

学習用データセットについて2回学習させてきました。ネットワークが全てを学習したのかをチェックする必要があります。  
ネットワークの出力するクラスが正しいのかどうかをチェックし、それが正解と一致しているのかをみます。  
最初に、テストデータを表示してみます。

```python
dataiter=iter(testloader)
images,labels=dataiter.next() # 次のミニバッチを読み込む

# 画像を表示する
print(images.shape,labels.shape)
imshow(torchvision.utils.make_grid(images))
print("GroundTruth: "," ".join("%5s"%classes[labels[j]] for j in range(4)))
```

先ほどの学習済みモデルを持ってきます。

```python
net=Net()
net.load_state_dict(torch.load(PATH)) # パラメータを設定する
```

それでは推論してみます。

```python
outputs=net(images)
print(outputs.shape)
print(outputs)
```

出力は10クラスのそれぞれの予測度です。一番値の高いクラスを予測クラスとします。

```python
_,predicted=torch.max(outputs,1) # 各クラスの中で値が一番高いクラスを予測クラスとする
print(predicted.shape)
print("Predicted: "," ".join("%5s"%classes[predicted[j]] for j in range(4)))
```

それではデータセット全体で見てみましょう。

```python
correct=0 # 正しく分類できた数
total=0 # 分類した数

with torch.no_grad():
    for data in testloader:
        images,labels=data
        outputs=net(images) # ネットワークの出力
        _,predicted=torch.max(outputs.data,1) # クラスを予測
        total+=labels.size(0) # データ数        
        correct+=(predicted==labels).sum().item() # 正しく予測できたデータ数
        
print("Accuracy of the network on the 10000 test images: %d %%"%(100*correct/total))
```

チャンスレートは10%です。  
予測できたクラスとそうでないクラスを見てみましょう。

```python
class_correct=list(0. for i in range(10)) # 正しく予測できたクラス
class_total=list(0. for i in range(10)) # 予測したクラス

with torch.no_grad():
    for data in testloader:
        images,labels=data
        outputs=net(images) # ネットワークの出力
        _,predicted=torch.max(outputs,1) # 確率→クラス
        c=(predicted==labels).squeeze()
        
        for i in range(4):
            label=labels[i] # ラベル
            class_correct[label]+=c[i].item()
            class_total[label]+=1
            
# 各クラスについて表示
for i in range(10):
    print("Accuracy of %5s : %2d %%"%(classes[i],100*class_correct[i]/class_total[i]))
```

表示できました。
