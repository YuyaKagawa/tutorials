## 世界各地

今回世界中の色んな場所でやってみます。  
なお場所の選び方は参考にした記事そのままです。

```python
locations=["国会議事堂","วัดพระแก้ว","New York City",
          "осударственный Эрмитаж","مكة المكرمة"]
locations
```

それぞれの緯度経度を取得します。  
今後のために、データフレームの形式で格納しておきます。

```python
lonlat_list=[] # 緯度経度のリスト

for l in locations: # 各場所について
    result=gmaps.geocode(l) # 地理情報の取得の結果
    lat=result[0]["geometry"]["location"]["lat"] # 緯度
    lon=result[0]["geometry"]["location"]["lng"] # 経度
    
    lonlat_list.append([lon,lat]) # リストに追加
    
R=pd.DataFrame(np.array(lonlat_list),columns=["lon","lat"],index=locations) # データフレームに格納、列名=["lon","lat"]、インデックス=場所名
R
```

目的の経度・緯度情報が得られました。

# 地図画像の取得

それでは前節で得た緯度経度情報からその周辺の地図画像を取得、表示をします。  
前節を実行し終えたことを前提に進めます。  

流れとしては以下の通りです。  
<ol>
    <li>Google MapのURLを設定します。なおこの部分に地図の各種設定が含まれます。</li>
    <li>URL先から画像を取得します。</li>
    <li>画像を一旦保存します。</li>
    <li>画像を表示します。</li>
</ol>

なおURLの部分で設定できる地図の種類"maptype"には以下の選択肢があります。  
<ul>
    <li><code>roadmap</code>: 道路や建物などが表示される、Google Mapのデフォルトのマップ</li>
    <li><code>satellite</code>: Google Earthの衛星画像</li>
    <li><code>hybrid</code>: <code>roadmap</code>と<code>satellite</code>を複合した地図</li>
    <li><code>terrain</code>: <code>roadmap</code>から建物などの情報を除いた地形情報のみのマップ</li>
</ul>

今回は<code>hybrid</code>とします。  
まずはURLの大まかな部分を定義します。

```python
url1="https://maps.googleapis.com/maps/api/staticmap?center=" # （緯度、経度）を設定
url2="&maptype=hybrid" # 地図の種類を設定
url3="&size=" # ピクセル数を設定
url4="&sensor=false" # GPSの情報を利用する際だけTrue
url5="&zoom=" # 地図の縮尺を設定
url6="&markers=" # マーカーの位置を設定
url7="&key=" # APIキーを設定
```

そして設定は下の通りにします。

```python
path_output="./maps/" # 画像の出力先
pixel="640x480" # 写真のピクセル数（解像度）
scale="18" # 航空写真の縮尺度合い
```

それでは、各場所ごとに地図画像を得ます。

```python
for place,lonlat in R.iterrows(): # Rの各行について
    loc="{},{}".format(lonlat["lat"],lonlat["lon"]) # URLの緯度経度情報部分作成
    url=url1+loc+url2+url3+pixel+url4+url5+scale+url6+loc+url7+api_key # 最終的なURLを作成
    
    path_image=path_output+"{}.png".format(place) # 画像ファイルのパス
    
    try:
        data=urllib.request.urlopen(url).read() # データ取得
        
        with open(path_image,mode="wb") as f: 
            f.write(data) # ファイル書き込み
    except urllib.error.URLError as e: # もしURLが存在しなかったら
        print(e) # エラーメッセージ表示
```

これで保存できたはずです。  
試しに画像を読み込んでみましょう。  

## 地図画像確認

```python
for place in R.index: # 各場所について
    print(place) # 場所名表示
    
    path_image=path_output+"{}.png".format(place) # 画像のパス 
    display_png(Image(path_image)) # jupyterで画像表示
```

確かに地図画像を得られました。
