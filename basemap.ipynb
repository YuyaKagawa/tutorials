BasemapはGSHHG(Global Self-consistent, Hierarchical, High-resolution Shoreline)、
オープンな海岸線データセットとGMT(Generic Mapping Tools)の川や州、国の境界線のデータセットを持っているので、地図上に境界線を描くことができます。  
例えば下のようなメソッドを持ちます。  
<ul>
    <li><code>drawcoastlines()</code>: 海岸線を引きます。</li>
    <li><code>fillcontinents()</code>: 大陸を色付きで表示します。
        ただ、特定の状況下では間違えて海の方を色塗りしてしまう可能性があります。
        その場合は色を指定して陸と水を塗る下のdrawlsmask()を使って上塗りするとよいようです。</li>
    <li><code>drawcountries()</code>: 国の境界線を描きます。</li>
    <li><code>drawstates()</code>: 北アメリカの州の境界線を描きます。</li>
    <li><code>drawrivers()</code>: 川を描きます。</li>
</ul>
上記の方法で海岸線と境界線を引く他に、画像を地図として扱うこともできます。例えば下のようなオプションがあります。
<ul>
    <li><code>drawlsmask()</code>: 画像として陸と海をそれぞれの特定の色で描きます。
        海岸線の描き方はいくつかあります。</li>
    <li><code>bluemarble()</code>: NASAのBlue Marble（宇宙からの航空写真）を地図の背景に指定できます。</li>
    <li><code>shadedrelief()</code>: Shaded Relief（陰影起伏図、鳥瞰図とか鳥の見る風景っぽい）を地図の背景に指定できます。</li>
    <li><code>etopo()</code>: etopo（アメリカ地球物理データセンターの提供するデータセット）を地図の背景に指定できます。<br>
        <a href="http://ofgs.aori.u-tokyo.ac.jp/~okino/gmtscripts/datainfo.html">
            http://ofgs.aori.u-tokyo.ac.jp/~okino/gmtscripts/datainfo.html</a>によると、定番中の定番らしいです。</li>
    <li><code>warpimage()</code>: 任意の画像を背景に指定できます。全地球をカバーしている必要があります。
        日付変更線を一番東に、南極点を一番北にする必要があります。</li>
</ul>

# 海岸線を描く

参考:  
Drawing the first map,  
https://basemaptutorial.readthedocs.io/en/latest/first_map.html

## 海岸線のみ

```python
m=Basemap(
    lon_0=0., # 中心の経度
    lat_0=0. # 中心の緯度
)
m.drawcoastlines() # 海岸線を描く
```

地球は球体で、二次元に投影する方法が様々あります。  
別の方法で投影してみましょう。  
ついでに陸と海を色分けしてみましょう。

## 海岸線&陸と海（色付き）

```python
m=Basemap(
    projection="ortho", # 宇宙から見た感じ 
    lat_0=0.,
    lon_0=0.
         )
m.drawmapboundary(fill_color="aqua") # 地球の端の線を描き、色を塗る
m.fillcontinents(color="coral", # 陸を色塗る
                lake_color="aqua" # 湖は改めて色塗る
                )
m.drawcoastlines() # 海岸線を描く
```

# 点をプロットする

参考:  
Drawing a point in a map,  
https://basemaptutorial.readthedocs.io/en/latest/basic_functions.html#drawing-a-point-in-a-map

<code>Basemap.plot()</code>を使って、地図上に点をプロットできます。

```python
m=Basemap(
    projection="ortho",
    lat_0=0.,
    lon_0=0.)
m.drawmapboundary(fill_color="aqua")
m.fillcontinents(color="coral",lake_color="aqua")
m.drawcoastlines()

x,y=m(0.,0.)

# マーカーをプロット
m.plot(x,y, # 座標はBasemapオブジェクトで
       marker="D", # マーカーの形 D: ダイヤモンド
       color="red", # マーカーの色
       markersize=10) # マーカーのサイズ
```

点をプロットするときは座標を指定しましょう。

ところで2つ以上の点をプロットするときは、<code>scatter()</code>を使いましょう。  
点の間の線を引きたいときは上のまま<code>plot()</code>を使いましょう。

```python
m=Basemap(
    projection="ortho",
    lat_0=0.,
    lon_0=0.)
m.drawmapboundary(fill_color="aqua")
m.fillcontinents(color="coral",lake_color="aqua")
m.drawcoastlines()

lons=[0.,10.,-20.,-20.] # 経度をリストの形で与える
lats=[0.,-10.,40.,20.] # 緯度をリストの形で与える

x,y=m(lons,lats)

# マーカーをプロット
m.scatter(x,y, # 座標はBasemapオブジェクトで
       marker="D", # マーカーの形 D: ダイヤモンド
       color="red", # マーカーの色
       ) # マーカーのサイズ
```
